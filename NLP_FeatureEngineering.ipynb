# Text Preparation: Remove stop words

```python
#Remove stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
import re
```

```python
REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]"')
BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
STOPWORDS = set(stopwords.words('english'))

def text_prepare(text):
    """
        text: a string
        
        return: modified initial string
    """
    
    text = text.lower()
    text = re.sub(REPLACE_BY_SPACE_RE," ",text)
    text = re.sub(BAD_SYMBOLS_RE,"",text)
    text = ' '.join([word for word in text.split() if word not in STOPWORDS])
    
    return text
```

```python
#Testing function
text_prepare("Captain Barbossa, long believed to be dead, has come back to life and is headed to the edge of the Earth with Will Turner and Elizabeth Swann. But nothing is quite as it seems")
```

```python
ls_train = [text_prepare(x) for x in train_df['overview']]
#X_val = [text_prepare(x) for x in X_val]
#X_test = [text_prepare(x) for x in X_test]
```

```python
#Join dataframe with a new column: Overview without stop words
train_df['overview_sw']=ls_train
```

# Tokenize, Vectorize and TF-IDF

```python

from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer()
vec.fit(train_df["overview_sw"])
transformed = vec.transform(train_df["overview_sw"])
text_features = pd.DataFrame(transformed.todense())
text_features.columns = vec.get_feature_names()
```

```python
#Testing vectorizer
text_features.loc[:, "assassin"]
```

```python
#Final dataframe with vectorized words
final_df=pd.concat([train_df, text_features], axis=1)
```
