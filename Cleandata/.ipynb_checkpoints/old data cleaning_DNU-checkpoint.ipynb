```python
#dependencies

import os
import csv
import pandas as pd
import numpy as np
```

```python
# Store filepath in a variable
movie_csv = "movies_testdata.csv"

movie_df = pd.read_csv(movie_csv)
```

```python
#split release date into month & year, add new columns

months = []
years = []
release_dates = movie_df["release_date"]
    
for x in range(0, len(movie_df)):
    split_date = release_dates[x].split("-")
    #print (split_date)
    months.append(split_date[1])
    years.append(split_date[0])

print(months)    
print(years)

movie_df['month'] = months
movie_df['year'] = years

movie_df
```
